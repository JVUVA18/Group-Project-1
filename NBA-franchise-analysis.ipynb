```python
import csv
import pandas as pd
import string
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
file_path = 'NBA-Complete.csv'

NBA_df = pd.read_csv(file_path)
NBA_df.fillna(0)
```

```python
x = NBA_df['Team']
y1 = NBA_df['Pct (2017)']
y2 = NBA_df['Attend. (2017)']

figure(figsize=(600, 200), dpi=20)
plt.rc('font', size=10)

# Create the bar graph
fig, ax1 = plt.subplots()
ax1.bar(x, y1, color='y')
ax1.set_xlabel('Teams')
ax1.set_ylabel('Win %', color='k')

# Create the second set of y-axis labels
ax2 = ax1.twinx()
ax2.stem(x, y2)
ax2.set_ylabel('Attendance', color='k')

ax1.set_xticklabels(x, rotation=90)

# Show the plot
plt.show()
```
